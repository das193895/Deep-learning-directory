```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## define the required function

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def swish(x, beta):
    return x * sigmoid(beta * x)

def swish_derivative(x, beta):
    return sigmoid(beta * x) + x * beta * sigmoid(beta * x) * (1 - sigmoid(beta * x))
```

```python
# Input values
x_values = np.linspace(-5, 5, 1000)

# Different values of beta
betas = [0.1, 1.0, 10]

# Create separate plots for each beta value
for beta in betas:
    plt.figure(figsize=(8, 6))
    plt.plot(x_values, swish(x_values, beta), label='Swish')
    plt.plot(x_values, swish_derivative(x_values, beta), label='Swish Derivative')
    plt.plot(x_values, np.maximum(0, x_values), label='ReLU', linestyle='--')
    plt.title(f'Swish Activation Function and its Derivative (beta={beta})')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.show()
```
