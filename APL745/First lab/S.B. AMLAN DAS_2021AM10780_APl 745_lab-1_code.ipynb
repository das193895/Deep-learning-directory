### 1.1 Function for data_preprocessing

```python
## Function for loading the data and train test split

'''
--> Enter the name of the file as a string for the variable -- "csv_file" 
--> Enter the name of the "target variable" as a string for the variable -- "target_variable"
--> train_size , test_size -- The ratio you want to split the data (80 ,20)

'''

def data_preprocessing(csv_file,train_size,test_size,The_target_variable):
    
    # load the data using pandas
    
    df = pd.read_csv(csv_file)
    
    print(f"The shape of your dataset is : {df.shape}")
    
    total_no_of_rows = df.shape[0]
    total_no_of_columns = df.shape[1]
    
    # The input features(dependent features) & The target variable(independent feature)
    
    X = df.drop(The_target_variable,axis = 1)
    y = df[[The_target_variable]]
    
    # Train_test_split
    
    no_of_rows_for_training = total_no_of_rows*(train_size/100)
    no_of_rows_for_testing = total_no_of_rows*(test_size/100)
    
    train_X = X.iloc[0:int(no_of_rows_for_training),:] 
    test_X = X.iloc[int(no_of_rows_for_training):total_no_of_rows,:]
 
    train_y = y.iloc[0:int(no_of_rows_for_training),:]
    test_y = y.iloc[int(no_of_rows_for_training):total_no_of_rows,:]
    
    print(f"The shape of your traing data set is: {train_X.shape}")
    print(f"The shape of your testing data set is : {test_X.shape}")
    
    ## Convert the pandas dataframe to to Numpy array to do model building
    
    train_X_arr = train_X.to_numpy()
    train_y_arr = train_y.to_numpy()
    test_X_arr = test_X.to_numpy()
    test_y_arr = test_y.to_numpy()
    
    print("your dataframe has been splitted to a traing and testing dataset as well as it has been converted to numpy array for furthe preprocessing")
    
    return train_X_arr, train_y_arr, test_X_arr, test_y_arr
    
```

```python
## Doing a train test split in 80:20 ratio

train_X_arr, train_y_arr, test_X_arr, test_y_arr = data_preprocessing("beam_data_1.csv",80,20,"Max Deflection (mm)")
```

## 2. Implementing Linear Regression

### 2.1 Using the basis function -- phi(x) = x (y = w0 + w1x1 + w2x2 +......+ wpxp)

```python
'''
---> Computing the least square loss

'''

def computing_cost_fn(feature_matrix, target_var, weight_matrix):
    
    ones_column = np.ones((feature_matrix.shape[0], 1))
    X = np.hstack((ones_column, feature_matrix))
    y = target_var

    no_of_samples = X.shape[0]
    
    summation = 0.0;
    
    
    for x,y in zip(X, y):
        y_hat = np.dot(x,weight_matrix)
        summation += (y_hat - y) ** 2
        
    final_cost = summation / (no_of_samples * 2.0)

    return final_cost[0]
```

```python
## Implementing the " GRADIENT DESCENT ALGORITHM "

'''
The function is taking [feature_matrix (inputs) , target_var, weight_matrix (Intial weight matrix) ,learning_rate, 
max_iter (Maximum Iteration)] as inputs and outputting the updated weight matrix and a matrix containing costs in each iteration.

'''

def gradient_descent(feature_matrix, target_var, weight_matrix,learning_rate,max_iter):
    
    iteration = 0
    
    # define a matrix for storing the losses
    cost = np.zeros(((max_iter),1))
    
    # stackiing a columns of 1 to the feature matrix
    ones_column = np.ones((feature_matrix.shape[0], 1))
    X = np.hstack((ones_column, feature_matrix))
    
    no_of_samples = X.shape[0]
    
    
    while iteration < max_iter:
        
        gradient = np.zeros((X.shape[1],)).reshape(1,-1).T  # taking a initial gradient
        
        cost[iteration] = computing_cost_fn(feature_matrix, target_var, weight_matrix) # storing the cost into the cost matrix in every iteration
        
        # looping through all the dataset to calculate gradients and updating the weight matrix
        
        for x,y in zip(X, target_var):
            
            y_hat = np.dot(x,weight_matrix)
            
            gradient = gradient + ((y_hat[0] - y[0]) * x.reshape(1,-1).T)  # gradient calculation
            
        
               
        weight_matrix -= (learning_rate * (gradient)/no_of_samples) # updation of weight matrix
        
        
        
        # If we get cost convergence (cost is not changing much)
        if iteration > 0 and np.abs(cost[iteration - 1] - cost[iteration]) < 1e-5:
            
            print("cost convergence",cost[iteration])
            
            break
            
        
        iteration = iteration + 1  
        
        
    return cost , weight_matrix   ## Returns the matrix which stores the cost in each etaration
```

```python
gradient_descent(train_X_arr,train_y_arr,weights,0.00001,100)
```

```python
'''
Implementing the Linear regression function using the gradient descent algorithm and predecting the y values using the test data

'''
def linear_Regression(testing_features,testing_target_var,feature_matrix, target_var, weight_matrix,learning_rate,max_iter):
    
    ## compute the optimum weights using gradient_descent algorithm
    
    cost_matrix , updated_weight_matrix = gradient_descent(feature_matrix, target_var, weight_matrix,learning_rate,max_iter)
    
    # adding a column of ones to the testing feature matrix 
    ones_column = np.ones((testing_features.shape[0], 1))
    
    X_test = np.hstack((ones_column, testing_features))
    
    no_of_samples = X_test.shape[0]
    
    y_test = testing_target_var
    
    
    # predicting the new y values with updated weight matrix
    y_pred = np.dot(X_test,updated_weight_matrix)
        
    return y_pred,cost_matrix  ## return the predicted y_value vector and the cost matrix which stores cost in each iteration 
```

### 2.1.1 Ridge Regression

```python
## Compute the cost function for ridge regression

def computing_cost_fn_ridge(feature_matrix, target_var, weight_matrix):
    
    ones_column = np.ones((feature_matrix.shape[0], 1))
    X = np.hstack((ones_column, feature_matrix))
    y = target_var

    no_of_samples = X.shape[0]
    no_of_columns = X.shape[1]
    
    summation = 0.0
    
    weight_sum = 0.0
    
    
    for x,y in zip(X, y):
        y_hat = np.dot(x,weight_matrix)
        summation += (y_hat - y) ** 2
        
    for w in weight_matrix:
        weight_sum += w ** 2   ## updating weights
        
        
    final_cost = (summation / (no_of_samples * 2.0)) + (weight_sum / no_of_columns)  ## computing cost

    return final_cost[0]
```

```python
## Compute the cost using The Intial weight matrix
computing_cost_fn_ridge(train_X_arr, train_y_arr, weights)
```

```python
## Gradient descent algorithm with Ridge regression

def gradient_descent_ridge(feature_matrix, target_var, weight_matrix,learning_rate,max_iter):
    
    iteration = 0
    
    cost = np.zeros(((max_iter),1))  ## defining a initial cost matrix
    
    ones_column = np.ones((feature_matrix.shape[0], 1))
    
    X = np.hstack((ones_column, feature_matrix))
    
    no_of_samples = X.shape[0]
    
    no_of_columns = X.shape[1]
    
    
    while iteration < max_iter:
        
        gradient = np.zeros((X.shape[1],)).reshape(1,-1).T  ## defining a initial gradient matrix
        
        cost[iteration] = computing_cost_fn_ridge(feature_matrix, target_var, weight_matrix)
        
        for x,y in zip(X, target_var):
            
            y_hat = np.dot(x,weight_matrix)
            
            gradient = gradient + ((y_hat[0] - y[0]) * x.reshape(1,-1).T)  # gradient calculation
            
            
        weight_sum = 0.0
        
        for w in weights:
            
            weight_sum += w 
            
               
        weight_matrix -= (learning_rate * ((gradient /no_of_samples) + ((weight_sum)/(2.0*no_of_columns)) ))  # updation of weight matrix
        

        
        if iteration > 0 and np.abs(cost[iteration - 1] - cost[iteration]) < 1e-5:
            
            print("cost convergence",cost[iteration])
            
            break
            
        
        iteration = iteration + 1
        
        
    return cost , weight_matrix 
```

```python
def linear_Ridge_Regression(testing_features,testing_target_var,feature_matrix, target_var, weight_matrix,learning_rate,max_iter):
    
    ## compute the optimum weights using gradient_descent algorithm
    
    cost_matrix , updated_weight_matrix = gradient_descent_ridge(feature_matrix, target_var, weight_matrix,learning_rate,max_iter)
    
    
    ones_column = np.ones((testing_features.shape[0], 1))
    
    X_test = np.hstack((ones_column, testing_features))
    
    no_of_samples = X_test.shape[0]
    
    y_test = testing_target_var
    
    
    y_pred = np.dot(X_test,updated_weight_matrix)
        
    return y_pred , cost_matrix
```

```python
def computing_cost_fn_lasso(feature_matrix, target_var, weight_matrix):
    
    ones_column = np.ones((feature_matrix.shape[0], 1))
    X = np.hstack((ones_column, feature_matrix))
    y = target_var

    no_of_samples = X.shape[0]
    no_of_columns = X.shape[1]
    
    summation = 0.0
    
    weight_sum = 0.0
    
    
    for x,y in zip(X, y):
        y_hat = np.dot(x,weight_matrix)
        summation += (y_hat - y) ** 2
        
    for w in weight_matrix:
        weight_sum += abs(w)
        
        
    final_cost = (summation / (no_of_samples * 2.0)) + (weight_sum / no_of_columns)

    return final_cost[0]
```

```python
## A function for converting the feature matrix and the target variable to its corresponding log values 

def convert_input_log(X,y,test_X,test_y):

    no_of_samples_inX = X.shape[0]
    
    no_of_columns_inX = X.shape[1]
    
    final_matrix_X = np.ones((no_of_samples_inX,no_of_columns_inX))
    
    final_test_matrix_X = np.ones((test_X.shape[0],test_X.shape[1]))
    
    ## converting the training and test feature matrices to its corresponding logarithmic values
    ## First column
    final_matrix_X[:,0:1] = np.log(X[:, 0:1]).reshape(-1,1)
    final_test_matrix_X[:,0:1] = np.log(test_X[:, 0:1]).reshape(-1,1)
    
    ## Second column
    final_matrix_X[:,1:2] = np.log(X[:, 1:2]).reshape(-1,1)
    final_test_matrix_X[:,1:2] = np.log(test_X[:, 1:2]).reshape(-1,1)
    
    ## 3rd column
    final_matrix_X[:,2:3] = np.log(X[:, 2:3]).reshape(-1,1)
    final_test_matrix_X[:,2:3] = np.log(test_X[:, 2:3]).reshape(-1,1)
    
    ## fourth column
    final_matrix_X[:,3:4] = np.log(X[:, 3:4]).reshape(-1,1)
    final_test_matrix_X[:,3:4] = np.log(test_X[:, 3:4]).reshape(-1,1)
    
    ## fifth column
    final_matrix_X[:,4:5] = np.log(X[:, 4:5]).reshape(-1,1)
    final_test_matrix_X[:,4:5] = np.log(test_X[:, 4:5]).reshape(-1,1)
    
    
    no_of_samples_iny = y.shape[0]
    
    no_of_columns_iny = y.shape[1]
    
    final_matrix_y = np.ones((no_of_samples_iny,no_of_columns_iny))  ## converting the traing target variable
    
    final_test_matrix_y = np.ones((test_y.shape[0],test_y.shape[1]))
    
    
    final_matrix_y[:,0:1] = np.log(y[:, 0:1]).reshape(-1,1)          ## converting the testing target variable
    final_test_matrix_y[:,0:1] = np.log(test_y[:, 0:1]).reshape(-1,1)
    
    
    return final_matrix_X,final_matrix_y,final_test_matrix_X,final_test_matrix_y
```

### 2.3 Linear regression with Pytorch

```python
## With pytorch

# Set the seed for reproducibility (optional)
torch.manual_seed(42)

# Generate a random column vector with gradients enabled
random_weights = torch.full((6, 1), fill_value=0.6, requires_grad=True)

def forward(X):
    return X@random_weights   # calculate the X*W matrix


def loss(y,y_predicted):
    return((y-y_predicted)**2).mean()  ## calculating the mean squared loss
```

```python
def gradient_descent_torch(feature_matrix, target_var,weight_matrix,learning_rate,max_iter):
    
    iteration = 0  ## Initialise the iteration
    
    cost = torch.full((max_iter, 1), fill_value=0.0)  ## Initialise the cost matrix
    
    ## adding a colum of ones to the feature_matrix
    ones_column = np.ones((feature_matrix.shape[0], 1))
    
    X_i = np.hstack((ones_column, feature_matrix))
    
    X = torch.from_numpy(X_i).float()
    
    target_var_tensor = torch.from_numpy(target_var).float()  ## converting the numpy array to a pytorch tensor
    
    no_of_samples = X_i.shape[0]
    
    
    while iteration < max_iter:
        
        y_pred = forward(X) # predicting the y
        
        loss_v = loss(target_var_tensor,y_pred) # calculating the loss
          
        cost[iteration] = loss_v  # adding the loss to a cost matrix
        
        # gradientts backword pass
        
        loss_v.backward()
        
        # update weights
        
        with torch.no_grad():
            weight_matrix -= learning_rate*weight_matrix.grad
            
        # zero gradients
        weight_matrix.grad.zero_()
        
        # If cost is not changing much
        if iteration > 0 and abs(cost[iteration - 1] - cost[iteration]) < 1e-5:
            
            print("cost convergence")
            
            break
        
        iteration += 1
        
        
    return cost , weight_matrix   ## Returns the matrix which stores the cost in each etaration
```

```python
def linear_Regression_torch(testing_features,testing_target_var,feature_matrix, target_var, weight_matrix,learning_rate,max_iter):
    
    ## compute the optimum weights using gradient_descent algorithm
    
    cost_matrix , updated_weight_matrix = gradient_descent_torch(feature_matrix, target_var,weight_matrix,learning_rate,max_iter)
    
    # adding a column of ones to the test matrix
    ones_column = np.ones((testing_features.shape[0], 1))
    
    X_test = np.hstack((ones_column, testing_features))
    
    no_of_samples = X_test.shape[0]
    
    X_test_tensor = torch.from_numpy(X_test).float()  ## converting the numpy array to a pytorch tensor

    y_pred = X_test_tensor@updated_weight_matrix
  
    return y_pred,cost_matrix
```

### 2.4 Using the basis function log(y) = log(x) (log(y) =[logx1,logx2,logx3.............])

```python
## A function for converting the feature matrix and the target variable to its corresponding log values 

def convert_input_log(X,y,test_X,test_y):

    no_of_samples_inX = X.shape[0]
    
    no_of_columns_inX = X.shape[1]
    
    final_matrix_X = np.ones((no_of_samples_inX,no_of_columns_inX))
    
    final_test_matrix_X = np.ones((test_X.shape[0],test_X.shape[1]))
    
    ## converting the training and test feature matrices to its corresponding logarithmic values
    ## First column
    final_matrix_X[:,0:1] = np.log(X[:, 0:1]).reshape(-1,1)
    final_test_matrix_X[:,0:1] = np.log(test_X[:, 0:1]).reshape(-1,1)
    
    ## Second column
    final_matrix_X[:,1:2] = np.log(X[:, 1:2]).reshape(-1,1)
    final_test_matrix_X[:,1:2] = np.log(test_X[:, 1:2]).reshape(-1,1)
    
    ## 3rd column
    final_matrix_X[:,2:3] = np.log(X[:, 2:3]).reshape(-1,1)
    final_test_matrix_X[:,2:3] = np.log(test_X[:, 2:3]).reshape(-1,1)
    
    ## fourth column
    final_matrix_X[:,3:4] = np.log(X[:, 3:4]).reshape(-1,1)
    final_test_matrix_X[:,3:4] = np.log(test_X[:, 3:4]).reshape(-1,1)
    
    ## fifth column
    final_matrix_X[:,4:5] = np.log(X[:, 4:5]).reshape(-1,1)
    final_test_matrix_X[:,4:5] = np.log(test_X[:, 4:5]).reshape(-1,1)
    
    
    no_of_samples_iny = y.shape[0]
    
    no_of_columns_iny = y.shape[1]
    
    final_matrix_y = np.ones((no_of_samples_iny,no_of_columns_iny))  ## converting the traing target variable
    
    final_test_matrix_y = np.ones((test_y.shape[0],test_y.shape[1]))
    
    
    final_matrix_y[:,0:1] = np.log(y[:, 0:1]).reshape(-1,1)          ## converting the testing target variable
    final_test_matrix_y[:,0:1] = np.log(test_y[:, 0:1]).reshape(-1,1)
    
    
    return final_matrix_X,final_matrix_y,final_test_matrix_X,final_test_matrix_y
```
